## EDA

```python
titlefont = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

lablefont = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
```

```python
# What is the gender distribution of passengers?
plt.style.use("ggplot")
plt.pie(df["Gender"].value_counts() , autopct= "%.2f%%", startangle= 90)
plt.legend(df["Gender"].value_counts().index)
plt.title("Gender distribution of passengers" , titlefont)

plt.show()
```

```python
# What is the age distribution of passengers?
sns.histplot(df['Age'] , bins = 90)
plt.title("Age Distribution" , fontdict= titlefont , pad = 10)
plt.xlabel("Age" , fontdict= lablefont , labelpad= 15)
plt.ylabel("Count" , fontdict= lablefont , labelpad= 15)
plt.grid(False)
plt.show()
```

```python
# Which nationalities travel the most?
df.groupby("Nationality")['Passenger ID'].count().sort_values(ascending= False)[:10].plot(kind= "bar" , colormap="coolwarm")
plt.title("Top 10 nationalities travel the most" , fontdict= titlefont , pad = 10 , loc="center")
plt.xlabel("Nationality" , fontdict= lablefont , labelpad= 15)
plt.ylabel("Passenger Count" , fontdict= lablefont , labelpad= 15)
plt.xticks(rotation = 45)
plt.grid(False)
plt.show()
```

```python
from matplotlib import colormaps
print(colormaps)
```

```python
# What are the top 10 busiest airports based on departures?
import matplotlib.cm as cm 

colors = cm.winter(np.linspace(0, 1, 10))

df.groupby("Airport Name")['Passenger ID'].count().sort_values(ascending= False)[:10].plot(kind= "bar" , color= colors)
plt.title("Top 10 nationalities travel the most" , fontdict= titlefont , pad = 10 , loc="center")
plt.xlabel("Airport Name" , fontdict= lablefont , labelpad= 15)
plt.ylabel("Passenger Count" , fontdict= lablefont , labelpad= 15)
plt.xticks(rotation = 90)
plt.grid(False)
plt.show()
```

```python
# Which airport countries have the highest number of flights?

colors = cm.coolwarm(np.linspace(0, 1, 10))

df.groupby("Airport Country Code")['Passenger ID'].count().sort_values(ascending= False)[:5].plot(kind= "bar" , color= colors)
plt.title("Airport Countries with highest number of flights" , fontdict= titlefont , pad = 10 , loc="center")
plt.xlabel("Country Name" , fontdict= lablefont , labelpad= 15)
plt.ylabel("Passenger Count" , fontdict= lablefont , labelpad= 15)
plt.xticks(rotation = 90)
plt.grid(False)
plt.show()
```

```python
# How does passenger age vary across continents?
sns.boxplot(data= df , x = "Continents" , y = "Age" )
plt.title("Passenger Age vary across continents" , fontdict= titlefont , pad = 10)
plt.ylabel("Age" , fontdict= lablefont , labelpad= 15)
plt.xlabel("Continents" , fontdict= lablefont , labelpad= 15)
plt.xticks(rotation = 45)
plt.show()
```

```python
# Do certain airports have higher cancellation rates?

temp=df.groupby("Airport Name")['Flight Status'].value_counts().reset_index()
temp[temp['Flight Status']=='Cancelled'].sort_values(by = "count" , ascending= False)[:10]
# .plot(kind= "bar" , colormap ="coolwarm")
sns.barplot(data = temp[temp['Flight Status']=='Cancelled'].sort_values(by = "count" , ascending= False)[:10] , x = "Airport Name" , y = "count")
plt.xticks(rotation = 90)
plt.xlabel("Airport Name" , fontdict= lablefont , labelpad= 15)
plt.ylabel("Cancellation Count" , fontdict= lablefont , labelpad= 15)
plt.title("Airports have highest cancellation rates" , fontdict= titlefont)
plt.show()
```

```python
# What is the average age of passengers for each flight status?
# How does passenger age vary across continents?
sns.boxplot(data= df , x = "Flight Status" , y = "Age" )
plt.title("Passenger Age vary across Flight Status" , fontdict= titlefont , pad = 10)
plt.ylabel("Age" , fontdict= lablefont , labelpad= 15)
plt.xlabel("Flight Status" , fontdict= lablefont , labelpad= 15)
plt.xticks(rotation = 45)
plt.show()
```
